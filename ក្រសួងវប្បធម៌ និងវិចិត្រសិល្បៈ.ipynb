```python
import os
import requests
from bs4 import BeautifulSoup

# Base URL of the page containing PDF links
base_url = "http://www.mcfa.gov.kh/site/listsdoc/40/"

# Create a directory to store the downloaded PDFs
os.makedirs("pdfs", exist_ok=True)

# Function to scrape PDF links from a given URL
def scrape_pdf_links(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find all <a> elements with an href attribute that ends with ".pdf"
    pdf_links = [a["href"] for a in soup.find_all("a", href=True) if a["href"].endswith(".pdf")]
    
    return pdf_links

# Iterate through multiple pages (adjust the range as needed)
for page_number in range(1, 3):  # Scraping pages 1 to 5 as an example#9
    page_url = f"{base_url}{page_number*10}" 
    
    # Scrape PDF links from the current page
    pdf_links = scrape_pdf_links(page_url)
    
    # Download each PDF from the current page
    for pdf_url in pdf_links:
        pdf_filename = os.path.join("pdfs", os.path.basename(pdf_url))

        # Download the PDF
        pdf_response = requests.get(pdf_url)
        with open(pdf_filename, "wb") as pdf_file:
            pdf_file.write(pdf_response.content)

        print(f"Downloaded: {pdf_filename}")

print("All PDFs from all pages downloaded successfully.")
```
